```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
DATA_PATH = '../../data/combined_tripdata.csv'
data = pd.read_csv(DATA_PATH)
print(data.head())
```

```python
# Check for non-standard datetime formats in the 'started_at' column
data['started_at'] = data['started_at'].astype(str)  # Ensure it's a string for processing
data['ended_at'] = data['ended_at'].astype(str)
```

```python
# Identify problematic rows
invalid_start_times = data[~data['started_at'].str.match(r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}(\.\d+)?', na=False)]
invalid_end_times = data[~data['ended_at'].str.match(r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}(\.\d+)?', na=False)]
# Display problematic rows
invalid_start_times.head(), invalid_end_times.head()
```

```python
# Convert timestamps while handling extra milliseconds
data['started_at'] = pd.to_datetime(data['started_at'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
data['ended_at'] = pd.to_datetime(data['ended_at'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
print(data.info())
```

```python
# Plot missing values before dropping
plt.figure(figsize=(10, 5))
data.isna().sum().plot(kind='bar', color='red', alpha=0.7)
plt.title("Missing Values in Each Column (Before Dropping)")
plt.xlabel("Columns")
plt.ylabel("Count of Missing Values")
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop rows with invalid datetime conversion (NaT values)
data = data.dropna(subset=['started_at', 'ended_at'])

# Verify the conversion
data[['started_at', 'ended_at']].head()
```

```python
# Remove rows where 'ended_at' is earlier than 'started_at' (data inconsistencies)
data = data[data['ended_at'] >= data['started_at']]
```

```python
# Compute ride duration in minutes
# Compute ride distance using the Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    lat1, lon1, lat2, lon2 = np.radians([lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

data['ride_duration'] = (data['ended_at'] - data['started_at']).dt.total_seconds() / 60
data['ride_distance_km'] = data.apply(lambda row: haversine(row['start_lat'], row['start_lng'], row['end_lat'], row['end_lng']), axis=1)
data['start_hour'] = data['started_at'].dt.hour
data['start_dayofweek'] = data['started_at'].dt.dayofweek
```

# Anomaly Detection - Isolation Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Remove zero-distance rides
data = data[data['ride_distance_km'] > 0]

# Select a subset of data (10%) to speed up computation
#data_sample = data
data_sample = data.sample(frac=1, random_state=42)

# Select features for anomaly detection
features = ['ride_duration', 'ride_distance_km']
X = data_sample[features]

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Display processed dataset shape
X_scaled.shape
```

```python
# Step 2: Train the Isolation Forest Model
iso_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
iso_forest.fit(X_scaled)

# Predict anomalies (-1 indicates anomaly, 1 indicates normal)
anomaly_predictions = iso_forest.predict(X_scaled)

# Compute anomaly scores (lower scores indicate higher anomaly likelihood)
anomaly_scores = iso_forest.decision_function(X_scaled)

# Step 3: Evaluate and Visualize Anomalies

# Count detected anomalies
num_anomalies = np.sum(anomaly_predictions == -1)
anamoly_threshold = np.percentile(anomaly_scores, 5)
print(f"Total Anomalies Detected: {num_anomalies}")
print(f"Anomaly Threshold: {anamoly_threshold}")

# Plot distribution of anomaly scores
plt.figure(figsize=(10, 5))
sns.histplot(anomaly_scores, bins=50, kde=True, color='blue')
plt.axvline(anamoly_threshold, color='red', linestyle='dashed', label="Anomaly Threshold")
plt.title("Isolation Forest Anomaly Score Distribution")
plt.xlabel("Anomaly Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Convert to structured dataset
data_features = {
    'ride_duration': data_sample['ride_duration'],
    'ride_distance_km': data_sample['ride_distance_km'],
    'start_lat': data_sample['start_lat'],
    'start_lng': data_sample['start_lng'],
    'end_lat': data_sample['end_lat'],
    'end_lng': data_sample['end_lng']
}
# Feature addition order
feature_order = ['ride_duration', 'ride_distance_km', 'start_lat', 'start_lng', 'end_lat', 'end_lng']

# Store metrics for comparison
metrics_history = []
num_anomalies_list = []

# Iterate over adding features one by one
for i in range(1, len(feature_order) + 1):
    selected_features = feature_order[:i]

    # Prepare dataset with selected features
    X_selected = np.column_stack([data_features[feature] for feature in selected_features])

    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_selected)

    # Train Isolation Forest model
    iso_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
    iso_forest.fit(X_scaled)

    # Predict anomalies
    anomaly_predictions = iso_forest.predict(X_scaled)
    anomaly_scores = iso_forest.decision_function(X_scaled)

    # Count anomalies
    num_anomalies = np.sum(anomaly_predictions == -1)
    num_anomalies_list.append(num_anomalies)

    # Store the mean anomaly score for comparison
    mean_anomaly_score = np.mean(anomaly_scores)
    std_anomaly_score = np.std(anomaly_scores)

    # Compute basic evaluation metrics (since we don’t have true labels, using pseudo-ground truth)
    precision = precision_score([1] * len(X_scaled), anomaly_predictions, pos_label=-1, zero_division=0)
    recall = recall_score([1] * len(X_scaled), anomaly_predictions, pos_label=-1, zero_division=0)
    f1 = f1_score([1] * len(X_scaled), anomaly_predictions, pos_label=-1, zero_division=0)

    metrics_history.append({
        'features_used': i,
        'feature_list': selected_features,
        'num_anomalies': num_anomalies,
        'mean_anomaly_score': mean_anomaly_score,
        'std_anomaly_score': std_anomaly_score,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    })
    print(f'mean_anomaly_score: {mean_anomaly_score}, std: {std_anomaly_score}')

    #Visualization of anomaly score distribution at each step
    plt.figure(figsize=(8, 4))
    sns.histplot(anomaly_scores, bins=50, kde=True, color='blue')
    plt.axvline(np.percentile(anomaly_scores, 5), color='red', linestyle='dashed', label="Anomaly Threshold")
    plt.title(f"Anomaly Score Distribution ({', '.join(selected_features)})")
    plt.xlabel("Anomaly Score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```

```python
metrics_df = pd.DataFrame(metrics_history)

# Plot the number of detected anomalies as features are added
plt.figure(figsize=(10, 5))
plt.plot(metrics_df['features_used'], metrics_df['num_anomalies'], marker='o', linestyle='-', color='red', label="Anomalies Detected")
plt.xlabel("Number of Features Used")
plt.ylabel("Number of Anomalies Detected")
plt.title("Impact of Feature Addition on Anomaly Detection")
plt.xticks(metrics_df['features_used'], metrics_df['feature_list'], rotation=45, ha='right')
plt.legend()
plt.show()

# Plot precision, recall, and F1-score over feature additions
plt.figure(figsize=(10, 5))
plt.plot(metrics_df['features_used'], metrics_df['precision'], marker='o', linestyle='-', label="Precision", color='blue')
plt.plot(metrics_df['features_used'], metrics_df['recall'], marker='o', linestyle='-', label="Recall", color='green')
plt.plot(metrics_df['features_used'], metrics_df['f1_score'], marker='o', linestyle='-', label="F1 Score", color='purple')
plt.xlabel("Number of Features Used")
plt.ylabel("Metric Value")
plt.title("Model Performance with Feature Addition")
plt.xticks(metrics_df['features_used'], metrics_df['feature_list'], rotation=45, ha='right')
plt.legend()
plt.show()

# Plot mean anomaly score and its standard deviation over feature additions
plt.figure(figsize=(10, 5))
plt.plot(metrics_df['features_used'], metrics_df['mean_anomaly_score'], marker='o', linestyle='-', label="Mean Anomaly Score", color='blue')
plt.fill_between(metrics_df['features_used'], 
                 metrics_df['mean_anomaly_score'] - metrics_df['std_anomaly_score'], 
                 metrics_df['mean_anomaly_score'] + metrics_df['std_anomaly_score'], 
                 color='blue', alpha=0.2, label="±1 Std Dev")
plt.xlabel("Number of Features Used")
plt.ylabel("Anomaly Score")
plt.title("Anomaly Score Evolution with Feature Addition")
plt.xticks(metrics_df['features_used'], metrics_df['feature_list'], rotation=45, ha='right')
plt.legend()
plt.show()
```
